```python
import pandas as pd

# Re-defining df for independent execution if previous cell was not run or kernel restarted
data = {
    'user_id': [1, 1, 2, 2, 3, 3, 3, 4, 4, 5],
    'item_id': [101, 102, 101, 103, 102, 104, 105, 101, 105, 102],
    'rating': [5, 3, 4, 5, 2, 4, 3, 5, 4, 5]
}
df = pd.DataFrame(data)

# Create a user-item interaction matrix
user_item_matrix = df.pivot_table(index='user_id', columns='item_id', values='rating')

print("\nUser-Item Interaction Matrix:")
display(user_item_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Fill NaN values with 0 for similarity calculation (or consider other strategies)
# For cosine similarity, missing values can be treated as 0 or handled differently.
# For simplicity, we'll fill with 0, implying no rating.
user_item_matrix_filled = user_item_matrix.fillna(0)

# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix_filled)

# Convert to a DataFrame for better readability, with user_ids as index and columns
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

print("\nUser Similarity Matrix (Cosine Similarity):")
display(user_similarity_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Re-defining df for independent execution if previous cell was not run or kernel restarted
data = {
    'user_id': [1, 1, 2, 2, 3, 3, 3, 4, 4, 5],
    'item_id': [101, 102, 101, 103, 102, 104, 105, 101, 105, 102],
    'rating': [5, 3, 4, 5, 2, 4, 3, 5, 4, 5]
}
df = pd.DataFrame(data)

# Prepare the data for splitting: we need a flat list of user-item-rating interactions
ratings = df[['user_id', 'item_id', 'rating']]

# Split the data into training and testing sets
# We use a test_size of 0.2 (20% for testing) and random_state for reproducibility
train_data, test_data = train_test_split(ratings, test_size=0.2, random_state=42)

print("Training data shape:", train_data.shape)
print("Test data shape:", test_data.shape)

display(train_data.head())
display(test_data.head())
```

```python
def predict_rating_user_based(user_id, item_id, user_item_matrix, user_similarity_df, k=3):
    # If the item or user is not in our matrix, we cannot make a prediction
    if item_id not in user_item_matrix.columns or user_id not in user_item_matrix.index:
        return np.nan

    # Get the row of similarities for the target user
    target_user_similarities = user_similarity_df.loc[user_id]

    # Get the ratings of other users for the target item
    item_ratings = user_item_matrix[item_id]

    # Combine similarities and item ratings, dropping NaN ratings for the item
    similarities_and_ratings = pd.concat([target_user_similarities, item_ratings], axis=1)
    similarities_and_ratings.columns = ['similarity', 'rating']

    # Filter out users who haven't rated the item or are the target user itself
    similarities_and_ratings = similarities_and_ratings.drop(user_id, errors='ignore')
    similarities_and_ratings = similarities_and_ratings.dropna()

    if similarities_and_ratings.empty:
        return np.nan

    # Sort by similarity and take the top k most similar users
    top_k_similar_users = similarities_and_ratings.sort_values(by='similarity', ascending=False).head(k)

    # Calculate weighted average
    if top_k_similar_users['similarity'].sum() == 0:
        return np.nan # Avoid division by zero

    predicted_rating = (top_k_similar_users['similarity'] * top_k_similar_users['rating']).sum() / top_k_similar_users['similarity'].sum()

    return predicted_rating

print("Prediction function defined.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Re-defining df and train/test data for independent execution if previous cells were not run or kernel restarted
data = {
    'user_id': [1, 1, 2, 2, 3, 3, 3, 4, 4, 5],
    'item_id': [101, 102, 101, 103, 102, 104, 105, 101, 105, 102],
    'rating': [5, 3, 4, 5, 2, 4, 3, 5, 4, 5]
}
df = pd.DataFrame(data)
ratings = df[['user_id', 'item_id', 'rating']]
train_data, test_data = train_test_split(ratings, test_size=0.2, random_state=42)

# Create user-item matrix from training data
train_user_item_matrix = train_data.pivot_table(index='user_id', columns='item_id', values='rating')

# Fill NaN values with 0 for similarity calculation on the training matrix
train_user_item_matrix_filled = train_user_item_matrix.fillna(0)

# Calculate user similarity based on training data
train_user_similarity = cosine_similarity(train_user_item_matrix_filled)
train_user_similarity_df = pd.DataFrame(train_user_similarity, index=train_user_item_matrix.index, columns=train_user_item_matrix.index)

# Make predictions for the test set
predictions = []
actual_ratings = []

for index, row in test_data.iterrows():
    user_id = row['user_id']
    item_id = row['item_id']
    actual_rating = row['rating']

    predicted_rating = predict_rating_user_based(user_id, item_id, train_user_item_matrix, train_user_similarity_df, k=3)

    if not np.isnan(predicted_rating):
        predictions.append(predicted_rating)
        actual_ratings.append(actual_rating)

# Calculate RMSE
if len(predictions) > 0:
    rmse = np.sqrt(mean_squared_error(actual_ratings, predictions))
    print(f"RMSE for User-Based Collaborative Filtering: {rmse:.4f}")
else:
    print("Could not make any predictions on the test set.")
```

```python
def generate_recommendations(user_id, user_item_matrix, user_similarity_df, k=3, num_recommendations=5):
    # Get items the user has already rated
    rated_items = user_item_matrix.loc[user_id].dropna().index.tolist()

    # Get all unique items
    all_items = user_item_matrix.columns.tolist()

    # Identify unrated items for the target user
    unrated_items = [item for item in all_items if item not in rated_items]

    if not unrated_items:
        print(f"User {user_id} has rated all available items or there are no unrated items.")
        return pd.Series(dtype=float)

    # Predict ratings for unrated items
    predictions = {}
    for item_id in unrated_items:
        predicted_rating = predict_rating_user_based(user_id, item_id, user_item_matrix, user_similarity_df, k=k)
        if not np.isnan(predicted_rating):
            predictions[item_id] = predicted_rating

    if not predictions:
        print(f"Could not predict ratings for any unrated items for User {user_id}.")
        return pd.Series(dtype=float)

    # Convert predictions to a Series and sort them
    predictions_series = pd.Series(predictions)
    top_recommendations = predictions_series.sort_values(ascending=False).head(num_recommendations)

    return top_recommendations

print("Recommendation generation function defined.")
```

```python
# Example: Generate recommendations for User 1
target_user = 1
recommendations = generate_recommendations(target_user, user_item_matrix, user_similarity_df, k=3, num_recommendations=5)

print(f"\nTop 5 Recommendations for User {target_user}:")
display(recommendations)
```
