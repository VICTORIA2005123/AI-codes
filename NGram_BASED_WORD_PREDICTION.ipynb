NGram BASED WORD PREDICTION





```python
#importing libraries
import nltk
from nltk.util import ngrams
from collections import Counter, defaultdict
import random
```

```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful"""
```

```python
nltk.download('punkt_tab')
```

```python
#tokenize the sentences
tokens=nltk.word_tokenize(corpus.lower())
```

```python
#building the ngram model
n=2
bigrams=list(ngrams(tokens,n))
```

```python
bigrams
```

```python
#creating the probability table
model=defaultdict(Counter)
for w1,w2 in bigrams:
  model[w1][w2]+=1
```

```python
model
```

```python
#funtion to predict the next word
def predict_next_word(word):
  if word in model:
    next_word=model[word]
    prediction=max(next_word,key=next_word.get)
    return prediction
  else:
    return "No prediction found"
```

```python
input_word="like"
print(predict_next_word(input_word))
```
